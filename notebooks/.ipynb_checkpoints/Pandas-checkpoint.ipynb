```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)

header = ["hash", "updateDate", "img", "link", "subscriberCount", "tagline", "title"]
df = pd.read_csv("./dump.csv", names=header)
```

```python
df = df[(df["updateDate"] >= "2021-10-11" ) &(df["updateDate"] <= "2021-10-12")]
```

```python
#existing rankings
res = df.groupby("link").agg({"subscriberCount": [np.min, np.max, np.ptp]}).reset_index()
res = res.sort_values([("subscriberCount", "ptp")], ascending=False)
res.head(10)
```

```python
# new comers ranking
occurance = df["link"].value_counts(ascending=True).to_frame().reset_index(level=0)
occurance = occurance[occurance["link"] == 1]
xs = list(occurance["index"])
new_comers = df.link.isin(xs)
new_comers = df[new_comers].sort_values("subscriberCount", ascending=False)
new_comers.head(10)
```
